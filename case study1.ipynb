# 3. Plot a bar graph of male vs female buyers participated in the sales.

```python
data['Buyer Gender'].value_counts()
```

```python
sns.set(font_scale=1.4)
data['Buyer Gender'].value_counts().plot(kind='bar', figsize=(7, 6),color=['red', 'lightgreen'], rot=0)
plt.xlabel("Gender", labelpad=14)
plt.ylabel("Count of People", labelpad=14)
plt.title("Count of People Who participated in the sales by Gender", y=1.02);
```
